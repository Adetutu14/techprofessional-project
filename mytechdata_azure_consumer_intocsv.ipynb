```python
#Let's import our libraries
from confluent_kafka import Consumer
import json
import pandas as pd
from datetime import datetime
from adlfs import AzureBlobFileSystem

```

```python
storage_account_name = "your_storage_account_name"
container_name = "your_container_name"
account_key = "your_account_key"
```

```python
# let's do our kafka configuration and append into a list we can create a dataframe

kafka_conf = {    
'bootstrap.servers': 'your_bootstarp.server', 
'security.protocol': 'SASL_SSL', 

'sasl.mechanism': 'PLAIN', 

'sasl.username': 'your sasl username', 
'sasl.password': 'your sasl_password',
'group.id':'your_data_group',
'auto.offset.reset': 'earliest'

}

kafka_topic = "tech_data"

consumer = Consumer(kafka_conf)
consumer.subscribe([kafka_topic])
print(f"The tech data to be consumed: {kafka_topic}")

try:
    techdata = []
    while True:
        msg = consumer.poll(1.0)
        if msg is None:
            continue
        if msg.error():
            print(f"Kafka is yet to receive the message: {msg.error()}")
            continue
        data = json.loads(msg.value().decode('utf-8'))
        print(f"tech data receieved from kafka: {data}")
        techdata.append(data)

except Exception as e:
    print('The error is: ', e)
finally:
    consumer.close()
```

```python
df = pd.DataFrame(techdata)
df.head()
```

```python
# Here is to save data in ADLS store in csv and append when new file comes in from the UI app

file = AzureBlobFileSystem(account_name=storage_account_name, account_key=account_key)

# Save directly to ADLS
csv_path = f"{container_name}/tech_data.csv"

#Check if the file exists in ADLS
if file.exists(csv_path):
    
    with file.open(csv_path, "r") as f:  #Read existing CSV
        existing_df = pd.read_csv(f)

        df = pd.concat([existing_df, df], ignore_index=True)   #Append new data

#Save back to ADLS
with file.open(csv_path, "w") as f:
    df.to_csv(f, index=False)

print(f"Uploaded CSV to ADLS: {csv_path}")
```
